```python
#query data via postgresql
query1 = """
ALTER TABLE death ALTER total_deaths TYPE int USING total_deaths::int;

SELECT d.start_date, d.end_date, s.state, d.place_of_death, d.covid_19_deaths, d.total_deaths FROM death d
	JOIN state_abbr s
	ON d.state = s.state
	WHERE d.place_of_death = 'Total - All Places of Death'
	AND d.start_date = '2020-01-01T00:00:00.000'
	AND d.end_date = '2020-12-31T00:00:00.000'
	ORDER BY total_deaths DESC;
"""
```

```python
#return query resultï¼šwhich state has the highest death in 2020?
results1 = connection.execute(query1).fetchall()

column_names1 = results1[0].keys()

for_research_question_1_df = pd.DataFrame(results1, columns = column_names1)

for_research_question_1_df
```

```python
#query data via postgresql
query_3 = """
SELECT v.mmwr_week, s.state, v.distributed_janssen, v.distributed_moderna, v.distributed_pfizer FROM vaccination v
	JOIN state_abbr s
	ON v.location = s.abbreviation
	WHERE v.mmwr_week = '48'
	AND v.location = 'NY'
	ORDER BY date DESC
	LIMIT 1;
"""
```

```python
# query data via postgresql
query_7 = """
select location,
	   round(avg(cast(covid_19_deaths as int)),2) as avg_covid_19_death,
	   round(avg(cast(total_deaths as int)),2) as avg_total_death,
	   round(avg(cast(series_complete_yes as int)),2) as avg_series_complete
from (select * 
	  from state_abbr s
	  join vaccination v on v.location = s.abbreviation
	  join death d on d.state = s.state) as t 
where mmwr_week = '48'
and place_of_death = 'Total - All Places of Death'
group by location;
"""
## With the constraint of No.48 mmwr_week and 'Total - All Places of Death', the query finds out the average death caused by covid-19, average total death, and average completed vaccine series on each state.


```

```python
results_7 = connection.execute(query_7).fetchall()

column_names_7 = results_7[0].keys()

for_research_question_7_df = pd.DataFrame(results_7, columns = column_names_7)

for_research_question_7_df
```
